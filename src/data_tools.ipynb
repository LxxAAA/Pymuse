```python
import csv
import pandas as pd
import nltk
```

```python
df = pd.read_csv("/Users/macbookpro/Desktop/music-info-retrieval/data/songdata.csv")
```

```python
lyrics = df['text'].iloc[0]
```

```python
print(lyrics)
```

```python
l = lyrics.split()
```

```python
m = ' '.join(l)
print(m)
```

```python
df['text'].iloc[0] = m
```

```python
df['text'].iloc[0]
```

```python
df.head(2)
```

```python
sentence = m
```

```python
tokens = nltk.word_tokenize(sentence)
```

```python
tokens
```

```python
tagged = nltk.pos_tag(tokens)
tagged[0:6]
```

```python
from nltk.draw.tree import TreeView
entities = nltk.chunk.ne_chunk(tagged)
# entities
TreeView(entities)._cframe.print_to_file('output.ps')
```
